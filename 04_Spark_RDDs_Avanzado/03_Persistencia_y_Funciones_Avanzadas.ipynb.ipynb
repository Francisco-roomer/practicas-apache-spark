# Almacenamiento en Cache

```python
!pip install findspark

import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Almacenamiento en Cache Parte 4").getOrCreate()
sc = spark.sparkContext
```

```python
rdd = sc.parallelize([item for item in range(10)])
```

```python
from pyspark.storagelevel import StorageLevel
```

```python
#queremos en este caso, solo que persista a nivel de memoria
rdd.persist(StorageLevel.MEMORY_ONLY)
```

dato interesante : al cambiar su nivel de persistencia debemos tener en cuenta que primero es necesario hacerle un unpersist() si no nos va a devolver un error

```python
rdd.unpersist()
```

```python
rdd.persist(StorageLevel.DISK_ONLY)
```

Dependiendo en que situacion estemos debemos escoger uno u otro nivel de persistencia con la idea de optimizar nuestras aplicaciones

```python
rdd.unpersist()
```

```python
# el sabe que con esta funcion de cache el nivel de persistencia simplemente es memoria
rdd.cache()
```
