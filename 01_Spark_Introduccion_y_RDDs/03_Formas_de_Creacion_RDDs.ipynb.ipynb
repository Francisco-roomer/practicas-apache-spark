## RDDs

```python
import sys
!{sys.executable} -m pip install findspark

import findspark

findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
# Instanciar el spark Context - RDD
sc = spark.sparkContext
```

```python
# Crear un RDD vacío
rdd_vacio = sc.emptyRDD
```

```python
# Crear un RDD con parallelize (vamos a crearlo con 3 particiones)
rdd_vacio3 = sc.parallelize([], 3)
```

```python
# Como verificar cuantas particiones tiene ( Es un RDD vacio cuyo num de particiones es 3 )
rdd_vacio3.getNumPartitions()
```

```python
# Crear RDD paralelizado que contiene datos (en esta caso una lista) podemos colocarle el num de particiones en este caso no tiene
rdd = sc.parallelize([1,2,3,4,5])
```

```python
# Para ver el contenido por el momento con la funcion collect
rdd.collect()
```

```python
# Crear un RDD desde un Archivo de Texto (Colocar la Ruta del archivo)
rdd_texto = sc.textFile('/content/rdd_source.txt')
```

```python
rdd_texto.collect()
```

```python
# la diferencia es que va a crear un archivo de texto donde es un solo registro
rdd_texto_completo = sc.wholeTextFiles("/content/rdd_source.txt")
```

```python
# En la primera columna el nombre del archivo y en la segunda el contenido
rdd_texto_completo.collect()
```

```python
# Creamos un RDD a partir de otro ya creado (existente) con una funcion lambda

rdd_suma = rdd.map(lambda x: x+1)

rdd_suma.collect()
```

```python
# Creamos un RDD a partir de un DataFrame

df = spark.createDataFrame([(1, 'jose'), (2, 'juan'),(3, 'fran')], ['id', 'nombre'])
```

```python
df.show()
```

```python
df_2 = spark.createDataFrame([(1, "Pedro"), (2, "moises")], ["id", "nombre"])
```

```python
df_2.show()
```
