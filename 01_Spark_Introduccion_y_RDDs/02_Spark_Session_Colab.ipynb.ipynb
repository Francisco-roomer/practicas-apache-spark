## Configuración de la SparkSession: `.master()` y `.appName()`

El método `.master()` define el entorno en el que se ejecutará nuestra aplicación de Spark. Dependiendo de dónde vayas a correr tu código, su argumento cambiará:

* **En un clúster:** Se debe pasar como argumento el nombre o la dirección del administrador de recursos del clúster. Por lo general, en estos casos se utiliza **YARN** o **Mesos** (por ejemplo, `.master("yarn")`).
* **En entorno local:** Se utiliza el argumento `"local[X]"`. Al escribir `.master("local[*]")`, el asterisco (`*`) le indica a Spark que debe utilizar **todos los núcleos (*cores*) disponibles en el procesador (CPU)** de la máquina para ejecutar la sesión de forma paralela.

Ten en cuenta que en la configuración de Spark, el parámetro para darle un nombre a tu aplicación es `.appName("NombreDeTuApp")`, mientras que `.master()` se encarga exclusivamente de definir el entorno de ejecución. Ambos suelen encadenarse al crear la sesión.
