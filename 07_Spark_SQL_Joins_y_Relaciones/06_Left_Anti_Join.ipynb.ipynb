```python
departamentos = spark.read.parquet("./data/departamentos")
departamentos.show()
```

```python
empleados = spark.read.parquet("./data/empleados")
empleados.show()
```

```python
from pyspark.sql.functions import col
```

left_anti_join

este tipo de join le permite averiguar que fila de mi conjunto de datos de la izquierda no tienen filas coincidentes en el conjunto de datos de la derecha.

y el conjunto de datos resultante contendra solo las columnas del conjunto de datos de la izquierda

```python
empleados.join(departamentos, col("num_dpto") == col("id"), "left_anti").show()
```

Nos va a traer en el df de la izquierda en este caso empleados que no tienen su correspondiente en el df de la derecha ( Pedro ) num_dpto ( 0 ) nos esta indicando que empleados no estan asignados a un departamento

```python
##  como cambia el cruce si lo invertidmos

departamentos.join(empleados, col("num_dpto") == col("id"), "left_anti").show()
```

el departamento informatica no tiene asignado a ningun trabajador
