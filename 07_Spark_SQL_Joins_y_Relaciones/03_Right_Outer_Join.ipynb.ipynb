```python
!pip install findspark

import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Right_Outer_Join").getOrCreate()
```

```python
departamentos = spark.read.parquet("./data/Departamentos")
departamentos.show()
```

```python
empleados = spark.read.parquet("./data/Empleados")
empleados.show()
```

```python
from pyspark.sql.functions import col
```

## Right_Outer_Join

```python
empleados.join(departamentos, col("num_dpto") == col("id"), "right_outer").show()
```

es el inverso al Left Outer va a mantener todos los datos del df de la derecha es este caso el de (departamentos) y rellena con valores Null aquellos Ids de departamentos que no fueron encontrados en num_dpto correspondiente al df de empleados si observamos con detenimiento el id 35 que corresponde a informatica no lo tenemos en el df de empleados

```python
# otra alternativa

empleados.join(departamentos, col("num_dpto") == col("id"), "rightouter").show()
```

```python
# otra alternativa mas simple

empleados.join(departamentos, col("num_dpto") == col("id"), "right").show()
```
