```python
!pip install findspark

import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Full_Outer_Join").getOrCreate()
```

```python
departamentos = spark.read.parquet("./data/Departamentos")
departamentos.show()
```

```python
empleados = spark.read.parquet("./data/empleados")
empleados.show()
```

```python
from pyspark.sql.functions import col
```

hay una cosa que tenemos que tener en cuenta en este tipo de join. El comportamiento es el mismo que el de combinar los Join Left Outer y Right Outer

```python
empleados.join(departamentos, col("num_dpto") == col("id"), "full_outer").show()
```

El left outer mantenia todas las filas de la izquierda y me rellenaba con nulos todos los de la derecha que no fuero encontrados

si nos detenemos a observar en el df Empleados existe el num de departanmetno 0 pero no existe el Id 0 por lo tanto ese id se rellenara con Null en este caso al igual que el nombre_dpto



```python
# Otra alternativa

empleados.join(departamentos, col("num_dpto") == col("id"), "outer").show()
```

el resultado el del full outer join nos permite ver a que departamento esta asignado un empleado y que departamento tienen empleados  y tambien cuales no estan aasignados a un departamento y que departamentos no tienen empleados Ejemplo Pedro o el departamento de informatica
