```python
!pip install findspark

import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Left_Semi_Join").getOrCreate()
```

```python
departamentos = spark.read.parquet("./data/departamentos")
departamentos.show()
```

```python
empleados = spark.read.parquet("./data/empleados")
empleados.show()
```

```python
from pyspark.sql.functions import col
```

El comportamiento de este tipo de join es similar al inner join excepto que el conjunto de datos resultante no incluye las columnas del conjunto de datos de la derecha - (otra forma de pensar es  que su comportamiento es el opuesto a left_anti_join ) donde el conjunto de los datos resultantes contiene las filas conincidentes

```python
## Left Semi Join
```

## Left Semi Join

```python
empleados.join(departamentos, col("num_dpto") == col("id"), "left_semi").show()
```

nos esta trayendo todos los elementos del dataframe que tuvieron su coincidente o que evaluaron la coindicion True del dataframe de la derecha. Podemos obsercar que el unico registro que no nos esta mostrando es Pedro con el num de dpot 0 y esto es por que el num de dpto 0 no se encuentra en el id correspondiente a la columna id

Asi podemos responder la pregunta diga cuales son los trabajadores que estan asignados a un departamento

```python
# Otra Alternativa

empleados.join(departamentos, col("num_dpto") == col("id"), "leftsemi").show()
```
