```python
empleados = spark.read.parquet('./data/empleados/')
```

```python
departamentos = spark.read.parquet('./data/departamentos/')
```

```python
from pyspark.sql.functions import col
```

## Manejo de nombres de columnas duplicadas

```python
# agregamos una nueva columna con la funcion withColumn que ser igual a la columna id

depa = departamentos.withColumn('num_dpto', col('id'))
```

```python
depa.printSchema()
```

```python
empleados.printSchema()
```

```python
# Devuelve un error PORQUE NOS DICE QUE ES AMBIGUO

empleados.join(depa, col('num_dpto') == col('num_dpto'))
```

```python
# Forma correcta

df_con_duplicados = empleados.join(depa, empleados['num_dpto'] == depa['num_dpto'])
```

```python
df_con_duplicados = empleados.join(depa, empleados['num_dpto'] == depa['num_dpto'])
```

```python
# si hace el inner join por que yo le estoy informando que dataframe tiene que tomar en este caso es la columna ambigua

df_con_duplicados.printSchema()
```

```python
df_con_duplicados.select(empleados['num_dpto']).show()
```

```python
df2 = empleados.join(depa, 'num_dpto')
```
