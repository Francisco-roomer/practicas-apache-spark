```python
!pip install findspark

import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Left_Outer_Join").getOrCreate()
```

```python
empleados = spark.read.parquet("./data/Empleados")
empleados.show()
```

```python
departamentos = spark.read.parquet("./data/Departamentos")
departamentos.show()
```

```python
from pyspark.sql.functions import col
```

##  left_outer

```python
empleados.join(departamentos, col("num_dpto") == col("id"), "left_outer").show()

```

si observan contiene todos los elementos de la izquierda es decir del df (empleados) y a rellenado con Null aquellos elementos del df de la derecha donde no se encontro el valor (en este caso el numero de departamento 0)

```python
# Otra alternativa : es leftouter en vez de left_outer

empleados.join(departamentos, col("num_dpto") == col("id"), "leftouter").show()
```

```python
# Otra alternativa : left (es mucho mas simple y muccho mas facil)

empleados.join(departamentos, col("num_dpto") == col("id"), "left").show()
```
