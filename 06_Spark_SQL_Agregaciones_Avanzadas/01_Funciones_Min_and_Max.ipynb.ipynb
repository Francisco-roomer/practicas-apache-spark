```python
!pip install findspark

import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Min y Max").getOrCreate()
```

```python
vuelos = spark.read.parquet("./data/vuelos.parquet")
```

```python
vuelos.printSchema()
```

```python
from pyspark.sql.functions import min, max, col
```

```python
vuelos.select(
    min(col("AIR_TIME").alias("Tiempo minimo de vuelo")),
    max(col("AIR_TIME").alias("Tiempo maximo de vuelo"))
).show()
```

```python
# aca especificamente tendriamos que ver que unidades de tiempo se estan colocando en nuestro datasets para saber si son min u horas
```

```python
vuelos.select(
    min(col("AIRLINE_DELAY")),
    max(col("AIRLINE_DELAY"))
).show()
```
