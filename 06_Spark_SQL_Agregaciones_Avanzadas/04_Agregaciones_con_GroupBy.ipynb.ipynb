# Agrupación y Agregación por Columnas (`groupBy`)

Las agregaciones generalmente se realizan en conjuntos de datos que contienen una o más **columnas categóricas** (preferiblemente con una cardinalidad baja o moderada). Algunos ejemplos comunes de variables categóricas son: el género (sexo), la edad (o rangos de edad), el nombre de una ciudad o el nombre de un país.

A diferencia de las agregaciones globales que se aplican a todo el DataFrame completo, las agrupaciones nos permiten aislar subgrupos de datos para aplicar las funciones matemáticas de manera independiente dentro de cada uno de ellos.

Llevar a cabo una agregación combinada con una agrupación es un **proceso de dos pasos**:

---

## El Proceso de Dos Pasos

### Paso 1: Definir la Agrupación (`groupBy`)
El primer paso consiste en segmentar el DataFrame mediante la transformación **`groupBy("col1", "col2", ...)`**. Es en este punto donde se especifican cuáles serán las columnas categóricas que servirán como clave para organizar y clasificar las filas en subgrupos.

```python
# Ejemplo de sintaxis para iniciar una agrupación
df_agrupado = df.groupBy("province", "city")

````python
vuelos = spark.read.parquet("./data/vuelos.parquet")
````

````python
vuelos.printSchema()
````

````python
# Primero vamos a hacer una agrupacion y luego una agregacion
````

queremos saber los aeropuetos de origenes cuantos tenemos queremos agrupar por los aeropuetos de origenes y contar cuantos vuelos sucedieron por esos aeropuertos

````python
from pyspark.sql.functions import desc
````

````python
vuelos.groupBy("ORIGIN_AIRPORT").count().orderBy(desc("count")).show()
````

````python
## AGRUPAR POR VARIAS COLUMNAS

vuelos.groupBy("ORIGIN_AIRPORT", "DESTINATION_AIRPORT").count().orderBy(desc("count")).show()
````
