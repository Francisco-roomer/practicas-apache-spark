```python
!pip install findspark

import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("sum y sumDistinct avg").getOrCreate()
```

```python
vuelos = spark.read.parquet("/content/data/vuelos.parquet")
```

```python
from pyspark.sql.functions import sum, sumDistinct, avg, count
```

## Funcion sum()

```python
# La funcion sum calcula la suma de los valores en una columna numerica

vuelos.printSchema()
```

```python
vuelos.select(sum("DISTANCE").alias("Suma de la distancia")).show()
```

## sumDistinct()

```python
# Suma solo los valores distintos de una columna (Si dos vuelos recorrieron la misma distancia y esta repetido el sumDistinct solo los sumara una sola vez )

vuelos.select(sumDistinct("DISTANCE").alias("Suma de la distancia")).show()
```

## avg()

```python
# avg Calcula el valor promedio de una columna numerica

# hicimos un promedio Manual y un promedio con la funcion avg

vuelos.select(
    avg("AIR_TIME").alias("promedio_aire"),
    (sum("AIR_TIME") / count("AIR_TIME")).alias("promedio_calculado")
).show()
```
