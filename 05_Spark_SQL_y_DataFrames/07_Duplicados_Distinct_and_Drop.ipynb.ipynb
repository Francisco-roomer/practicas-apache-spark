# Eliminación de Duplicados: `distinct()` vs `dropDuplicates()`

Aunque ambas transformaciones se utilizan para limpiar registros repetidos en un DataFrame, existe una diferencia clave en su flexibilidad y comportamiento:

### 1. Comportamiento idéntico (Por defecto)
Si no especificamos ninguna columna, ambas funciones se comportan de la misma manera: Spark evaluará **todas las columnas** del DataFrame para determinar si una fila está duplicada y quedarse solo con los registros únicos.

### 2. La flexibilidad de `dropDuplicates()`
La gran diferencia es que `dropDuplicates()` nos permite especificar un **subconjunto de columnas** para aplicar la lógica de desduplicación.

> 💡 **En resumen:**
> * **`df.distinct()`**: Compara obligatoriamente *todas* las columnas del DataFrame.
> * **`df.dropDuplicates(["columna_A", "columna_B"])`**: Solo toma en cuenta las columnas indicadas para decidir si un registro es repetido, ignorando si el resto de las columnas tienen valores distintos.

```python
print(f"El conteo del DataFrame Original es {df.count()}")
print(f"El conteo del DataFrame sin duplicados es {df_sin_duplicados.count()}")
```

```python
# Funcion dropDuplicates
```

```python
dataframe = spark.createDataFrame([(1,"azul",567),(2,"Rojo",487),(1,"azul",345),(2,"verde",783)]).toDF("id","color","importe")
```

```python
dataframe.show()
```

```python
# si yo quisiera borrar aquellos registros de este DataFrame atendiendo la columna id y color

dataframe.dropDuplicates(["id","color"]).show()
```
