# Manejo de Datos Faltantes (*Missing Data*) en Spark

En el mundo real, los datos con los que trabajamos a menudo no están tan limpios como nos gustaría. Esto puede deberse a que los flujos de datos evolucionan con el tiempo, a errores en la captura o a integraciones de diferentes sistemas; el resultado es que algunas columnas tendrán valores y otras tendrán vacíos (`Null` o `None`).

Es de vital importancia abordar este problema al inicio de la manipulación de los datos. De lo contrario, nos expondremos a sorpresas desagradables, como que un pipeline de preprocesamiento de larga duración falle a mitad de camino debido a un valor nulo inesperado.

---

## La solución de Spark: El submódulo `df.na`

La comunidad de Spark reconoce que lidiar con datos faltantes es una necesidad cotidiana. Por ello, Spark proporciona una interfaz dedicada llamada **`DataFrameNaFunctions`**, a la cual podemos acceder fácilmente en cualquier DataFrame utilizando la propiedad **`df.na`**.

Las dos estrategias más habituales y efectivas para tratar los datos perdidos o incorrectos son:

### 1. Eliminar las filas con valores perdidos (`drop`)
Consiste en remover por completo los registros que tengan valores nulos en una, varias o todas sus columnas.

```python
# Elimina cualquier fila que contenga al menos un valor nulo
df_limpio = df.na.drop()

# Elimina la fila solo si tiene nulos en columnas específicas
df_limpio = df.na.drop(subset=["edad", "telefono"])


## Funcion Valores Nulos

````python
df = spark.read.parquet("./data")
````

````python
df.count()
````

````python
df.na.drop().count()
````

````python
# otr alternativa

df.na.drop("any").count()
````

````python
# otra alternativa

df.dropna().count()
````

Vamos a borrar ahora con un poco mas de control especificando que columnas

````python
# el parametro subset hace que sea mas focalizado en este caso en la columna "views"

df.dropna(subset=["views"]).count()
````

````python
df.dropna(subset=["views", "likes", "dislikes"]).count()
````

## Imputacion Rellenarlas con un valor determinado

````python
df.orderBy("views").select("views", "likes","dislikes").show()
````

````python
# fillna y el parametro value hace que se sustituyan todas las columnas que tenemos en nuestra consulta

df.fillna(value=0).orderBy("views").select("views", "likes","dislikes").show()
````

````python
# para rellenar especificamente con el parametro vlaue y subset podemos seleccionar que columnas

df.fillna(0,subset=["likes", "dislikes"]).orderBy("views").select("views", "likes","dislikes").show()
````
