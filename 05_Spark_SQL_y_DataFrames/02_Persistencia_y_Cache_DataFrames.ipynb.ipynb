# Persistencia y Almacenamiento en Caché de DataFrames

Al igual que con los RDDs, los DataFrames en Spark se pueden persistir o almacenar en caché utilizando los mismos niveles de almacenamiento (`MEMORY_ONLY`, `MEMORY_AND_DISK`, etc.). Sin embargo, existe una **gran diferencia y ventaja** al almacenar en caché un DataFrame en comparación con un RDD.

---

## La Ventaja del Motor Spark SQL

La optimización radica en que **Spark SQL conoce el esquema (*Schema*)** de los datos dentro de un DataFrame. Gracias a esto, en lugar de guardar los registros fila por fila como objetos Java planos (como hace un RDD), Spark realiza las siguientes optimizaciones de manera automática:

1. **Almacenamiento Columnar:** Organiza los datos en un formato orientado a columnas en la memoria.
2. **Compresión Inteligente:** Aplica algoritmos de compresión específicos según el tipo de datos de cada columna (por ejemplo, compresión para enteros, cadenas, etc.) para minimizar el uso de espacio.

[Image of Row-oriented vs Columnar data storage layout]

---

## Conclusión: DataFrame vs. RDD

> 📊 **Impacto en los Recursos:**
> Como resultado de la organización columnar y la compresión, **se requerirá muchísimo menos espacio en la memoria RAM** para almacenar un DataFrame en caché que para almacenar un RDD, incluso cuando ambos estén respaldados exactamente por el mismo archivo de datos.

Por esta razón, siempre que sea posible, es preferible utilizar DataFrames para optimizar el uso de memoria en tu clúster.

```python
# vamos a crear un df sencillo

df = spark.createDataFrame([(1 , "a"),(2 , "b"), (3 , "c"),(4 , "d")], ["id", "valor"])
```

```python
df.show()
```

```python
 # Dataframe persitencia de igual manera que los rdd y esto persistira nuesto dataframe en memoria

df.persist()
```

```python
# Podemos retirarlo de la memoria como un RDD

df.unpersist()
```

```python
# Para almecenarlo en memoria

df.cache()
```

```python
# almaacenar el nivel de Storage

from pyspark.storagelevel import StorageLevel
```

```python
df.persist(StorageLevel.MEMORY_AND_DISK)
```
