# Creación de DataFrames a partir de Fuentes de Datos

Las dos clases principales en Spark SQL para leer y escribir datos son **`DataFrameReader`** y **`DataFrameWriter`**.

Para leer datos, utilizamos una instancia de la clase `DataFrameReader`, la cual está disponible en la sesión de Spark a través del objeto **`spark.read`**.

---

## El Patrón Común de Lectura

El método estándar e integrado para interactuar con `DataFrameReader` sigue la siguiente estructura:

```python
spark.read.format(...).option("key", "value").schema(...).load()

### Componentes Principales al Leer Datos

| Elemento | ¿Es Opcional? | Descripción |
| --- | --- | --- |
| **`format`** | **No** | Define el tipo de archivo o fuente de datos que se va a leer. |
| **`option`** | **Sí** | Permite pasar configuraciones en formato clave-valor para modificar el comportamiento de la lectura. |
| **`schema`** | **Sí** (Depende) | Define la estructura de columnas y tipos de datos. Puede ser obligatorio o inferido según el formato. |

---

## Detalle de los Componentes

### 1. `format`

No es opcional. Puede ser una de las fuentes de datos integradas en Spark o un formato personalizado.

* Para **formatos integrados**, se puede usar un nombre corto como `json`, `parquet`, `orc`, `jdbc`, `csv` o `text`.
* Para **fuentes de datos personalizadas**, se debe proporcionar el nombre completo del paquete o clase externa.

### 2. `option`

`DataFrameReader` tiene un conjunto de opciones predeterminadas para cada formato de origen. Podemos anular estos valores por defecto proporcionando pares clave-valor específicos (por ejemplo, `.option("header", "true")` para archivos CSV).

### 3. `schema`

Algunas fuentes de datos ya tienen el esquema incrustado dentro de los propios archivos (como **Parquet, ORC y JDBC**); en estos casos, Spark infiere el esquema automáticamente. Para otros formatos basados en texto (como CSV o JSON), es una buena práctica (y a veces obligatorio) proporcionar un esquema definido manualmente para mejorar el rendimiento y evitar errores.

---

## Métodos Alternativos (Atajos de Lectura)

Spark ofrece métodos abreviados de forma directa para los formatos más comunes. Ambas formas presentadas a continuación son equivalentes:

### JSON

```python
spark.read.json("path")
# Es equivalente a:
spark.read.format("json").load("path")

```

### Parquet

```python
spark.read.parquet("path")
# Es equivalente a:
spark.read.format("parquet").load("path")

```

### JDBC (Bases de Datos Relacionales)

```python
spark.read.jdbc("url", "table", properties)
# Es equivalente a:
spark.read.format("jdbc").options(...).load()

```

### ORC

```python
spark.read.orc("path")
# Es equivalente a:
spark.read.format("orc").load("path")

```

### CSV

```python
spark.read.csv("path")
# Es equivalente a:
spark.read.format("csv").load("path")

```

### Text

```python
spark.read.text("path")
# Es equivalente a:
spark.read.format("text").load("path")

```

## CrearDataFrames a partir de fuentes de datos

````python
!pip install findspark

import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Dataframes A partir de datos ").getOrCreate()
sc = spark.sparkContext
````

Vamos a subir la carpeta Data que contiene varios formatos

````python
# Crear un DF a partir de texto

df = spark.read.text("./data/dataTXT.txt")
````

````python
# si quisieramos ver todo sin que se trunque la columna usamos el parametro truncate=False

df.show(truncate=False)
````

````python
# podemos observar que el nombre de las columnas lo ha omitido (queremos que se la primera fila)

df_csv = spark.read.csv("./data/dataCSV.csv")
df_csv.show()
````

````python
#para eso invovamos el parametro option y el header igual a true

df_csv1 = spark.read.option("header", "true").csv("./data/dataCSV.csv")
df_csv1.show()
````

# parte 2

veamos como se le puede cambiar el delimitador

````python
# leer archivo de texto con un delimitador diferente

df3 = spark.read.option("header", "true").option("delimiter", ",").csv("./data/dataCSV.csv")
df3.show()
````

````python
df_tab = spark.read.option("header", "true").option("delimiter", "|").csv("./data/dataTab.txt")
df_tab.show()
````

````python
# crear un DF a partir de la lectura de un archivo JSON pero ademas proporcionarle un Schema

from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DateType
````

````python
json_schema = StructType([
    StructField("Color", StringType(), True),
    StructField("edad", IntegerType(), True),
    StructField("fecha", DateType(), True),
    StructField("pais", StringType(), True)
])
````

````python
df4 = spark.read.schema(json_schema).json("./data/dataJSON.json")
df4.show()
````

````python
df4.printSchema()
````
