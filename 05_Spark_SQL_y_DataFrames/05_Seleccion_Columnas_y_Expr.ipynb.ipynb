# Manipulación y Operaciones Estructuradas en DataFrames

Una vez que sabemos crear DataFrames, el siguiente paso es manipularlos y estructurarlos utilizando las **operaciones estructuradas** que nos proporciona Spark.

A diferencia de los RDDs, las operaciones estructuradas están diseñadas para ser más **relacionales**. Esto significa que estas operaciones reflejan de manera directa el tipo de expresiones que se utilizan comúnmente en el lenguaje SQL, tales como filtrados, transformaciones, uniones (*joins*), agrupaciones, entre otras.

---

## Categorías de Operaciones: Transformaciones vs. Acciones

Al igual que ocurre con los RDDs, las operaciones estructuradas en DataFrames se dividen estrictamente en dos categorías, manteniendo exactamente la misma semántica:

### 1. Transformaciones (*Lazy Evaluation*)
Tienen una **evaluación perezosa** (*Lazy*). Esto significa que Spark no computa los datos de inmediato cuando defines una transformación. En su lugar, simplemente va registrando un plan de ejecución (el linaje de los datos) con los pasos que debe seguir.

### 2. Acciones (*Eager Evaluation*)
Tienen una **evaluación inmediata** (*Eager*). Son las operaciones que obligan a Spark a poner en marcha todo el motor de cálculo, optimizar el plan de ejecución registrado por las transformaciones y ejecutar la tarea real para devolver un resultado final o guardar los datos.

```python
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType

df.select(
    col('likes').cast(IntegerType()),
    col('dislikes').cast(IntegerType()),
    (col('likes').cast(IntegerType()) - col('dislikes').cast(IntegerType())).
alias('aceptacion')
).show()
```

```python
# as es lo mismo que Alias pero con esta alternativa selectExpr podemos acer una expresion tambien

df.selectExpr('likes', 'dislikes', '(CAST(likes AS INT) - CAST(dislikes AS INT)) as aceptacion').show()
```

```python
# Diferentes ecpresiones podemos contar los diferentes video ID y el alias sera videos

df.selectExpr("count(distinct(video_id)) as videos").show()
```
