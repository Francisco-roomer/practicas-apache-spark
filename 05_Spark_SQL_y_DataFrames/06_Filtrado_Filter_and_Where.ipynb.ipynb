```python
df = spark.read.parquet("/content/data/datos.parquet")

df.show()
```

## Funcion Filter

```python
# Filter

from pyspark.sql.functions import col
```

```python
df.filter(col("video_id") == "2kyS6SvSYSE").show()
```

```python
# Where (vamos a crear un dataframe a partir de la lectura de nuestros datos pero con una condicion de filtro)

df1 = spark.read.parquet("/content/data/datos.parquet")

df1.where(col("trending_date") != "17.14.11").show()
```

```python
df2 = spark.read.parquet("/content/data/datos.parquet")

df2.where(col("likes") > 5000)
```

```python
# como la podemos complicar un poco mas con filter con la comparacion &

df2.filter((col("trending_date") != "17.14.11") & (col("likes") > 7000)).show()
```

```python
# como lo podemos hacer de otra forma (aca podriamos colcar .filter().filter()) tastas veces como se quiera hasta que nos de el resultado requerido

df2.filter(col("trending_date") != "17.14.11").filter(col("likes") > 7000).show()
```
