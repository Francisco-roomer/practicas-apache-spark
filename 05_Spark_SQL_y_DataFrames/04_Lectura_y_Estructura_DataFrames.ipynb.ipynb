## Creacion de un DataFrames

```python
rdd = sc.parallelize([item for item in range(10)]).map(lambda x: (x, x** 2))
```

```python
rdd.collect()
```

```python
df  = rdd.toDF(["numero","cuadrado"])
```

### crear un schema a a partir de un rdd con schema

```python
# creamos un rdd a partir de tuplas ( ) pero estas se componen de 3 posiciones longitud 3 que tienen = - int - string - float

rdd1 = sc.parallelize([(1, "jose", 35.5),(2, "Francisco", 25.1),(3, "teresa", 55.5),(5, "Karina", 13.7),])
```

```python
# crear schema

from pyspark.sql.types import StructType, StructField, IntegerType, StringType, FloatType, DoubleType
```

```python
# via q de crear un schema - (se crea una estructura se indican que nombres de columnas, tipos de datos y si recibe datos NaN)

schema1 = StructType([
    StructField("id", IntegerType(), True),
    StructField("nombre", StringType(), True),
    StructField("sueldo", DoubleType(), True),
])
```

```python
# Via 2 mas senculla

schema2 = "`id` INT, `Nombre` STRING, `Saldo` DOUBLE"
```

```python
df1 = spark.createDataFrame(rdd1, schema=schema2)
```

```python
df1.printSchema()
```

```python
df1.show()
```

```python
df2 = spark.createDataFrame(rdd1, schema=schema1)
```

```python
df2.show()
```
