# La Transformación `limit()`

`limit()` es una transformación que devuelve un nuevo DataFrame tomando únicamente las primeras $N$ filas del DataFrame original.

Esta operación se utiliza comúnmente en combinación con un ordenamiento (`orderBy` o `sort`) para determinar de manera eficiente los $N$ elementos superiores (máximos) o inferiores (mínimos) de un conjunto de datos.

---

## Ejemplo de uso en PySpark

Imagina que tienes un DataFrame con calificaciones de estudiantes y quieres obtener los 3 mejores puntajes:

```python
# Ordenamos de forma descendente por la columna 'calificacion' y tomamos los 3 primeros
top_3_estudiantes = df.orderBy(df["calificacion"].desc()).limit(3)

# Mostramos el resultado (Acción)
top_3_estudiantes.show()

````python
# que pasa si de nuestro dataframe que ya esta ordenada queremos el top 10 views?

df.orderBy(col("views").desc()).limit(10).show()
````
