```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Ejercicio Parte 3").getOrCreate()
sc = spark.sparkContext
```

```python
importes = sc.textFile("/num.txt")
```

```python
importes.count()
```

```python
importes.min()
```

```python
importes.max()
```

```python
lista_top15 = importes.top(15)
```

```python
top_15 = sc.parallelize(lista_top15)
```

```python
top_15.saveAsTextFile("./data/salida/")
```

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    rdd = sc.parallelize(list(range(1, n + 1)))
    return rdd.reduce(lambda x,y: x * y)
```

```python
factorial(0)
```

```python
factorial(2)
```

```python
factorial(3)
```

```python
factorial(4)
```
