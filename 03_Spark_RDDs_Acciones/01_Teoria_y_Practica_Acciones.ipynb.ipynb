```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("RDD Acciones").getOrCreate()
```

```python
sc = spark.sparkContext
```

## Accion Funcion reduce()

```python
rdd = sc.parallelize([2,4,6,8])
```

```python
#debemos tener en cuenta que al ser una accion reduce va a desencadenar automaticamente el DAG por lo tanto nos va a devolver un resultado

rdd.reduce(lambda x,y : x+y)
```

Se sumaron todos los elementos que estaban en el RDD existente con una funcion lambda que en este caso necesita de 2 parametros x,y.

```python
rdd1 = sc.parallelize([1,2,3,4])
```

```python
#Cambiamos la funcion para multiplicar todos los elementos que estan en nuestor RDD

rdd1.reduce(lambda x,y : x*y)
```

# Accion count()

```python
rdd1 = sc.parallelize(["F","r","a","n","a","n","d","o","s"])
```

```python
# Importante count cuenta sin el 0 como en programacion el hace un conteo normal

rdd1.count()
```

```python
# una lista de un bucle for comprengation

rdd2 = sc.parallelize([item for item in range(10)])
```

```python
rdd2.count()
```

# Accion collect()

```python
rdd3 = sc.parallelize("Hoola francisco estas usando apache spark".split(" "))
```

```python
# Recolecta todos los elementos que estan en el RDD

rdd3.collect()
```

```python
rdd4 = sc.parallelize([(item, item **2) for item in range(20)])
```

```python
rdd4.collect()
```

# Acciones take, max y saveAsTextFile



# take()

```python
rdd_original = sc.parallelize("Hola Mundo este es un texto".split(" "))
```

```python
rdd_original.collect()
```

```python
# La funcion Take nos toma los primeros N elementos que encuentre dentro del RDD

rdd_original.take(2)
```

# max()

```python
# La funcion max nos devuelve el maximo de los numeros que encuentre en un RDD que este construido a partir de numeros

rdd_max = sc.parallelize([item / (item + 1) for item in range(10)])
```

```python
rdd_max.max()
```

```python
rdd_max.collect()
```

# saveAsTextFile


```python
rdd_original.collect()
```

```python
# se creo una carpeta que en este caso tiene dos ficheros
rdd_original.saveAsTextFile("./rdd")
```

```python
# Al combinar la disminucion de particion y salvarlo en una carpeta nos devolvio un rdd con una sola particion

rdd_original.coalesce(1).saveAsTextFile("./rdd1")
```

Esta tecnica es muy usada cuando no son ficheros muy grandes (bastante comun cuando vamos a utilizar ficheros en desarrollo para pruebas etc el comprimir una sola particion para tener un solo fichero a veces facilita mucho las acciones)
